# Definig and reading the input file

```python
def read_file(file):
  input_file=open(file)
  file=input_file.readlines()
  print(len(file))
  return file
```

#**Data Cleaning Function**

```python
def data_prep(some_list):
  file=some_list
  lines_list = []
  i=0
  for line in file:
    lines=line.split('),')
    for _ in lines:
      lines1= re.sub('[^a-zA-Z0-9, ]+', '', _)
      lines1=lines1.strip()
      lines1=lines1.replace(' ','')
      lines_list.insert(i,lines1)
      i=i+1
  return lines_list
```

# **Data Split Function**

```python
def split_lists(new_list,a):
  Split_list1=new_list[:a]
  Split_list2=new_list[a:]
  return Split_list1,Split_list2

```

# Mapper Functions


```python
def mapper1(List1):
  keyvalue=[]
  a_new_list=List1
  for list_vales in a_new_list:
    year, temp = list_vales[0:4],list_vales[7:10]
    #print(type(temp))
    #temp=int(temp)
    keyvalue.append([year,int(temp)])
  #print(year)
  #print(temp)
  return keyvalue

```

```python
def mapper2(List2):
  keyvalue1=[]
  a_new_list1=List2
  for list_vales1 in a_new_list1:
    year, temp = list_vales1[0:4],list_vales1[7:10]
    #print(type(temp))
    #temp=int(temp)
    keyvalue1.append([year,int(temp)])
  #print(year)
  #print(temp)
  return keyvalue1
```

# Sort Function

```python
def sortedlists(list1,list2):
  one_list=list1+list2
  one_list.sort(key= lambda x:x[0])
  return one_list
```

# Partition Function


```python
def partition_func(sorted_list1):
  partition1=[]
  partition2=[]
  sort_list=sorted_list1
#  partition1=sort_list[:1000]
#   partition2=sort_list[1000:]
  for values in sort_list:
    if values[0] in ['2010','2011','2012','2013','2014','2015']:
      partition1.append(values)
    else:partition2.append(values)
  return partition1,partition2
```

# Reducer Function

```python
def reducer_func(partition1):
  result_list=[]
  partition_list=partition1
  last_key,max_temp=(None,0)
  for year_temp in partition_list:
    year,temp=year_temp[0],year_temp[1]
    if last_key and last_key !=year:
      result_list.append([last_key,max_temp])
      (last_key,max_temp)=(year,temp)
    else:(last_key,max_temp)=(year,max(max_temp,temp))
    
  return result_list
    
```
